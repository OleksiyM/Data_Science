## Нейронна мережа

**Нейронна мережа** - це математична модель, яка імітує роботу нервової системи людини, зокрема мозку. Вона складається з великої кількості з'єднаних штучних нейронів, які співпрацюють для вирішення певних завдань.

Порівняно з мозком, нейронна мережа має кілька схожостей:

1. **Нейрони**: В нейронній мережі є штучні нейрони, які аналогічні нейронам у мозку. Кожен нейрон приймає вхідні сигнали, обробляє їх і передає вихідний сигнал іншим нейронам.

2. **Зв'язки між нейронами**: Нейрони у нейронній мережі з'єднані між собою зв'язками, які передають сигнали. Це аналогічно зв'язкам між нейронами у мозку, які передають електричні імпульси.

3. **Ваги зв'язків**: Кожен зв'язок у нейронній мережі має свою вагу, яка визначає важливість сигналу, що його пересилає. Це подібно до сили зв'язків між нейронами у мозку, де деякі зв'язки можуть бути сильнішими, ніж інші.

4. **Навчання**: Нейронні мережі можуть навчатися на основі вхідних даних. Під час навчання вони адаптують ваги зв'язків для оптимізації вихідного результату. Це подібно до того, як мозок навчається і робить висновки на основі досвіду.

5. **Вирішення задач**: Нейронні мережі використовуються для вирішення різноманітних завдань, таких як розпізнавання зображень, розпізнавання мови, прогнозування, управління роботами тощо. Це відбувається шляхом обробки вхідних даних через нейронну мережу та генерації вихідних результатів.

Зв'язуючи будову нейрона людини з будовою нейронної мережі, можна провести наступну аналогію:

1. **Дендрити** (вхідні з'єднання):

- У людському нейроні: Дендрити - це вхідні волокна нейрона, які приймають сигнали від інших нейронів або від нейронних вузлів у тілі клітини.
- У нейронній мережі: Вхідні з'єднання нейрона мережі - це аналогічні дендритам. Вони приймають вхідні сигнали від інших нейронів або від вхідних даних.
2. **Сома** (тіло клітини):

- У людському нейроні: Сома - це центральна частина клітини, де відбувається обробка сигналів, отриманих через дендрити.
- У нейронній мережі: Сома нейрона мережі відповідає обробці вхідних сигналів, вони об'єднують інформацію з різних вхідних з'єднань.
3. **Аксон** (вихідний зв'язок):

- У людському нейроні: Аксон - це вихідний зв'язок нейрона, який передає сигнали до інших нейронів через синапси.
- У нейронній мережі: Вихідний зв'язок нейрона мережі відповідає передачі сигналів до інших нейронів або до вихідного шару через синапси.
4. **Синапси** (з'єднання між нейронами):

- У людському нейроні: Синапси - це точки з'єднання між дендритами одного нейрона та аксонами інших нейронів, де відбувається передача сигналів.
- У нейронній мережі: Синапси відповідають зв'язкам між нейронами мережі, де відбувається передача вагованих сигналів від одного нейрона до іншого.

## Історія розвитку ідеї.

1. **Виникнення ідеї (1943-1958)**: Перші ідеї про штучні нейронні мережі виникли ще в 1940-х роках. Модель "нейронної мережі" була вперше описана в 1943 році МакКаллохом та Піттсом, які запропонували модель, що наслідувала роботу нейронів у мозку.

2. **Перші нейромережеві моделі (1958-1969)**: У 1958 році Френк Розенблатт розробив першу штучну нейронну мережу, відому як персептрон. Це був простий алгоритм класифікації, який став підґрунтям для подальшого дослідження.

3. **Період зменшення інтересу (1970-1980)**: У 1970-1980 роках інтерес до нейромережевих моделей спадав через обмежену потужність обчислювальних систем та обмежену доступність даних для навчання.

4. **Відродження (1980-1990)**: В 1980-х роках з'явилися нові методи навчання нейромереж, такі як алгоритм зворотного розповсюдження помилок, що сприяло відродженню інтересу до них.

5. **Етап розвитку (1990-2000)**: У 1990-х роках нейромережеві технології почали застосовуватися в різних областях, включаючи розпізнавання образів, обробку природної мови, прогнозування тощо.

6. **Зростання популярності (2000-нинішній час)**: З появою потужних обчислювальних ресурсів і великих обсягів даних, нейромережеві методи стали ще більш популярними. Вони застосовуються в різних сферах, включаючи медицину, фінанси, технології безпеки, машинне навчання тощо.

## Переваги і недоліки глибокого навчання

**Переваги класичного навчання**:
1. **Простота реалізації**: Класичні методи машинного навчання, такі як лінійна регресія або метод опорних векторів, зазвичай відносно прості для реалізації та розуміння.

2. **Ефективність на невеликих наборах даних**: В деяких випадках, коли у вас обмежені обсяги даних, класичні методи можуть працювати ефективніше, оскільки вони менше схильні до перенавчання.

3. **Легше пояснення результатів**: Результати класичних моделей часто легше інтерпретувати, що дозволяє розуміти, як саме модель приймає рішення.

**Переваги глибокого навчання**:

1. **Здатність працювати з складними даними**: Глибокі нейронні мережі можуть ефективно вирішувати завдання, які пов'язані зі складними даними, такими як зображення, аудіо або природна мова.

2. **Автоматичне відборення ознак**: Глибокі нейронні мережі можуть автоматично вивчати корисні ознаки з даних, що дозволяє їм ефективно вирішувати завдання без необхідності ручної інженерії ознак.

3. **Можливість вирішувати завдання, які важко формалізувати**: Глибоке навчання дозволяє вирішувати завдання, для яких складно визначити правила або формалізувати евристику, наприклад, розпізнавання облич, голосовий асистенти, автоматизоване перекладання тощо.

# Математична реалізація нейрона

![alt text](https://textbook.edu.goit.global/python/data-science-remaster/v1/img/module-09/Structure-of-Perceptron.png)

Загальна ідея

1. **Inputs(input layer)**: це набір фічей (колонок) які ми передаємо в глиб прихорваних слоїв (hidden layers) (або внутрішніх слоїв). Кожен $x$ --> ($x_1$, $x_2$, ..., $x_l$) це окрема фіча.

2. **Ініціалізуємо ваги (weights)** $w$ --> ($w_1$, $w_2$, ..., $w_l$). Тобто на скільки та чи інша фіча важлива, або на скільки зв'язок цих значень важливий.

3. **Пряме поширення сигналу (Forward Propagation)**.
- Після цього ми сумуємо всі значення в тілі нейрона (neyron) $N_1$ в прихорваному шарі 1. Додаємо зсув (bias) упередження про зсув данних або їх неідеальність $ net = \sum_{i=0}^{l}(x_iw_i) $.

- Рахуємо **активаційну функцію** (гіпотезу) $a = f(net)$ для нейрона $N_1$, тобто сумму нашого сигналу з минулого слою.

- **Обчислення втрат (Loss Computation)**: Після прямого поширення мережа порівнює прогнозовані вихідні значення з фактичними значеннями за допомогою функції втрат. Ця функція визначає, наскільки добре мережа вирішує поставлену задачу.

!!!
- **Зворотне поширення помилки (Backpropagation)**

!!!

- **Оновлення параметрів (Parameters Update)**: Оновлення ваг і зсувів проводиться за допомогою алгоритму оптимізації який використовує похідні втрат для керування оновленнями так, щоб мінімізувати втрати мережі.
  - Стохастичний градієнтний спуск (SGD)
  - Адаптивний градієнтний спуск (Adaptive Gradient Descent)

  Ці алгоритми адаптують швидкість навчання (learning rate) для кожного параметра на основі його історії градієнтів.
  - Adam (Adaptive Moment Estimation)
  
  Adam є комбінацією методів RMSprop та Momentum. Він використовує оцінки першого та другого моменту градієнту для адаптивного налаштування швидкості навчання для кожного параметра.

  - RMSprop (Root Mean Square Propagation)

  RMSprop адаптивно налаштовує швидкість навчання для кожного параметра шляхом ділення градієнту на середньоквадратичне значення попередніх градієнтів.

  - Adagrad (Adaptive Gradient Algorithm)

   Adagrad зберігає накопичену квадратичну суму градієнтів для кожного параметра та використовує її для адаптивного налаштування швидкості навчання.

  - Nesterov Accelerated Gradient (NAG)

  Цей метод використовує "миттєвий" градієнт, який вираховується в майбутньому позиції параметрів, що ближче до поточного положення.
  
  - AdaDelta

  Цей алгоритм подібний до RMSprop, але він використовує замість глобального середнього квадратичного значення швидкості навчання локальне середнє квадратичне значення зміненого швидкості навчання.


4. **Повторення кроків**: Ці кроки повторюються для кожного батча даних у навчальному наборі протягом кількох епох, доки функція втрат не буде зменшена до заданого рівня або до досягнення максимальної кількості епох навчання.





### Гіпотеза

![alt text](https://textbook.edu.goit.global/python/data-science-remaster/v1/img/module-09/multilayer-pepceptron.png)


Я якщо у прихованному шарі декілька нейронів?
Тоді запис стане більш навантаженним. Ось розшифрування для велечин вагів.


$$w_{ij}^{k}$$

i - номер аутпут нода (номер нода в шарі k, ЯКИЙ МИ ФОРМУЄМО)

j - номер інпут нода (номер нода, в шарі k-1, ЯКИЙ ДАЄ СВІЙ ВПЛИВ)

k - номер шару

Тоді всі записи будуть мати матричні види.
Тоді для l=3

Вхідні дані
$$
X = \begin{pmatrix}
x_{1}  \\
x_{2} \\
x_{3}
\end{pmatrix}
$$

Матриця вагів шара 1 для шару 2 буде мати вигляд

$$
W^{(1)} = \begin{pmatrix}
w_{11}^{(1)}  & w_{12}^{(1)}  & w_{13}^{(1)}  \\
w_{21}^{(1)}  & w_{22}^{(1)}  & w_{23}^{(1)}  \\
w_{31}^{(1)}  & w_{32}^{(1)}  & w_{33}^{(1)}
\end{pmatrix}
$$

Функція активації $g$ для нейрона i=1 шару k=2.  

$a_1^{2} = g(∑_{j=1}^3w_{i1}x_i)$


Тоді матриця функції активації 2 слоя має вигляд:

$$Z^{(2)} = W^{(1)}X$$

$$A^{(2)} = g(Z^{(2)})$$

А для третього слоя і далі:

$$Z^{(k)} = W^{(k-1)}A^{(k-1)}$$

В загальному вигляді, гіпотези має вигляд

$$h_w(x) = g(Z^{k})$$

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def h(weights, inputs):
    z = np.dot(weights, inputs)
    return sigmoid(z)
```

### Функції активації



1. Лінійна функція (Linear Activation):
$$ f(x) = x $$
2. Сигмоїда (Sigmoid)
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

3. Гіперболічний тангенс (Hyperbolic Tangent, Tanh):
$$ f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$
4. Softmax
$$\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$$
5. ReLU (Rectified Linear Unit):
$$f(x) = \max(0, x)$$
6. Leaky ReLU:
$$\text{LeakyReLU}(x) = \begin{cases} x, & \text{if } x > 0 \\ \alpha x, & \text{інакше} \end{cases}$$

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def softmax(x):
    exp_vals = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_vals / np.sum(exp_vals, axis=-1, keepdims=True)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

## Практика

```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model

# Побудова моделі
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Відображення архітектури моделі
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```
