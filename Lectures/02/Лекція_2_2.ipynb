## Лінійні графіки

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(12, 6),
           dpi=80
           )
date = pd.date_range(start='2021-09-01', freq='D', periods=8)
plt.plot(date, [23, 17, 17, 16, 15, 14, 17, 20])

plt.xlabel('Дата',
           fontsize='small',
           fontstyle ='normal',
           fontweight ='medium',
           color='midnightblue'
           )
plt.ylabel('Температура', fontsize='small', color='midnightblue')

plt.title('Денна погода у м. Полтава',
          loc = 'left',
          fontsize=15)

plt.text(date[0], 15, 'Осінь досить тепла', color="blue")

plt.show()
```

```python
# https://matplotlib.org/stable/gallery/color/named_colors.html
```

```python
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
```

```python
fig, axs = plt.subplots(figsize=(12, 6), dpi=80)
axs.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature', color = 'navy')
axs.plot(date, [19, 11, 16, 11, 10, 10, 11, 16], label='night temperature', color = 'orange')

plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')

plt.title('Температура в м. Полтава', fontsize=15)
plt.text(date[0], 15, 'Осінь досить тепла', color="blue")
plt.legend()

plt.show()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 6), dpi=80)

axs[0,0].plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
axs[1,1].plot(date, [19, 11, 16, 11, 10, 10, 11, 16], label='night temperature')

axs[0,0].set_title('Денна', fontsize=10)
axs[1,1].set_title('Нічна', fontsize=10)

fig.suptitle('Температура в м. Полтава', fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(12, 6), dpi=80)
plt.plot(
    date,
    [23, 17, 17, 16, 15, 14, 17, 20],
    label="day temperature",
    linestyle="--",
    color="#FF5733",
)
plt.plot(
    date,
    [19, 11, 16, 11, 10, 10, 11, 16],
    label="night temperature",
    linestyle="-.",
    color="#061358",
)
plt.xlabel("Дата", fontsize="small", color="midnightblue")
plt.ylabel("Температура", fontsize="small", color="midnightblue")
plt.title("Температура в м. Полтава", fontsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6), dpi=80)
plt.plot(
    date,
    [23, 17, 17, 16, 15, 14, 17, 20],
    label="day temperature",
    linestyle="--",
    color="#FF5733",
    linewidth=2,
    marker="D",
)
plt.plot(
    date,
    [19, 11, 16, 11, 10, 10, 11, 16],
    label="night temperature",
    linestyle=":",
    color="#061358",
    linewidth=2,
    marker="^",
)
plt.ylim(0, 30)
plt.xlabel("Дата", fontsize="small", color="midnightblue")
plt.ylabel("Температура", fontsize="small", color="midnightblue")
plt.title("Температура в м. Полтава", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

## Кругові діаграми


```python
plt.figure(figsize=(12, 6), dpi=80)
labels = [
    "Junior Software Engineer",
    "Senior Software Engineer",
    "Software Engineer",
    "System Architect",
    "Technical Lead",
]

data = [63, 31, 100, 2, 11]
explode = [0.15, 0.1, 0, 0, 0]
plt.pie(
    data,
    labels=labels,
    shadow=True,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)

plt.show()
```

## Діаграми в полярних координатах

```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2.0 * np.pi, 1000)
r = np.sin(6 * theta)

plt.polar(theta, r)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(projection="3d")

theta_max = 8 * np.pi
n = 1000
theta = np.linspace(0, theta_max, n)
x = theta
z = np.sin(theta)
y = np.cos(theta)
ax.plot(x, y, z, "g")

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(projection="3d")

x = [5, 10, 15, 20]
z = [10, 0, 5, 15]
y = [0, 10, 5, 25]
s = [150, 130, 30, 160]
ax.scatter(x, y, z, s=s)

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_wireframe(x, y, z)

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_surface(x, y, z)

plt.show()
```
