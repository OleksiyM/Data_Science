# Що таке рекурентні нейронні мережі?


**Рекурентні нейронні мережі (RNN)** є класом нейронних мереж, які спеціально призначені для роботи з послідовними даними, де кожен елемент в послідовності (наприклад, слово в реченні або часовий крок у часовому ряді) взаємодіє з попередніми елементами. Основним відмінним рішенням RNN є використання рекурентних зв'язків, які дозволяють інформації потікти через мережу з попередніх кроків часу.

У класичній RNN кожен часовий крок мережі приймає на вхід вектор вхідних даних та вектор попереднього стану (повернутий з попереднього кроку) і видає вектор вихідних даних та новий стан, який може бути використаний на наступному кроці. Основна ідея полягає в тому, що мережа має можливість "запам'ятовувати" попередні інформаційні стани та використовувати їх для врахування контексту при обробці нових даних.

Основний принцип роботи рекурентних нейронних мереж (RNN):

1. **Інпут даних.**

На початку роботи мережі на вхід надходять вхідні дані. Це може бути послідовність слів у тексті, часовий ряд, звуковий сигнал або будь-яка інша послідовність даних.

2. **Ініціалізація стану.**

Першим кроком роботи мережі є ініціалізація внутрішнього стану. Це може бути вектор нулів або будь-який інший початковий стан.

3. **Передача даних через мережу.**

На кожному часовому кроці RNN приймає вхідні дані та попередній стан і видає вихідні дані та новий стан. Цей процес виконується **рекурентно** для кожного елементу в послідовності.

4. **Обчислення внутрішніх значень.**

На кожному кроці мережа обчислює внутрішні значення, які включають в себе лінійну комбінацію вхідних даних та попереднього стану, а також застосування нелинійної функції активації. Це допомагає мережі вирішувати складні нелинійні залежності.

5. **Запам'ятовування стану.**

Один з ключових аспектів роботи RNN - це здатність запам'ятовувати інформацію з попередніх кроків. Це досягається за рахунок зв'язків між різними часовими кроками, які дозволяють мережі "згадувати" попередні значення.

6. **Вихідні дані і новий стан.**

На кожному кроці мережа генерує вихідні дані та новий внутрішній стан, який потім використовується наступним кроком.

## 1. Інпут даних. Токенізація

**Токенізація слова** - це процес розбиття тексту на окремі одиниці, які називаються "токенами". З математичної точки зору основна мета токенізації полягає в перетворенні тексту на послідовність числових значень або векторів, що можуть бути подані у вигляді даних, придатних для обробки моделями машинного навчання.

Види токенізацій

1. **Токенізація за словами (Word Tokenization):** Текст розділяється на слова. Цей підхід зазвичай використовується для роботи з природною мовою, де слова є основними одиницями.

2. **Токенізація за символами (Character Tokenization):** Текст розділяється на окремі символи. Цей підхід може бути корисним для деяких задач, де символи мають важливе значення, або якщо потрібно зменшити обсяг даних.

3. **Токенізація за підрядками (Substring Tokenization):** Текст розділяється на підрядки або фрази. Цей підхід може використовуватися для розбиття тексту на більш малий рівень, ніж слова, але більший рівень, ніж символи.

4. **Токенізація за звуками (Phoneme Tokenization):** Текст розділяється на фонеми, що є найменшими звуковими одиницями мови. Цей підхід часто використовується у розпізнаванні мовлення та синтезі мовлення.

5. **Токенізація за грамемами (Grammatical Tokenization):** Текст розділяється на граматичні одиниці, такі як слова, фрази, речення або абзаци. Цей підхід дозволяє аналізувати текст на різних рівнях складності.

6. **Токенізація за регулярними виразами (Regex Tokenization):** Текст розділяється за допомогою заданого регулярного виразу, який вказує на шаблон токенів. Цей підхід дозволяє гнучко визначати, як розділяти текст на токени.

**Кодування буквами (Character-Level Encoding):**

**Плюси:**
- Дозволяє врахувати більш дрібні деталі тексту, оскільки кожен символ розглядається окремо.
- Підходить для мовних завдань, де слова можуть бути незвичними або новими.

**Мінуси:**
- Зазвичай призводить до більш великої кількості токенів, особливо великих текстів, що може призвести до більшого обсягу пам'яті та обчислювальних ресурсів.



```python
from tensorflow.keras.preprocessing.text import Tokenizer

# Зразок тексту у форматі рядка
text = "Привіт, світ!"

# Ініціалізуємо Tokenizer та надаємо йому рядок тексту для навчання
tokenizer = Tokenizer(char_level=True)
tokenizer.fit_on_texts([text])

# Токенізуємо текст
char_tokens = tokenizer.texts_to_sequences([text])[0]

# Отримуємо словник токенів
char_index = tokenizer.word_index

# Виводимо результат
print("Текст:", text)
print("Токенізовані букви:", char_tokens)
print("Словник токенів:", char_index)
```

**Кодування словами (Word-Level Encoding):**

**Плюси:**
- Забезпечує більшу семантичну розуміння, оскільки слова є основними одиницями розуміння мови.
- Зазвичай дає меншу кількість токенів, особливо для тексів з обмеженим словником.

**Мінуси:**
- Може не враховувати деякі дрібні деталі тексту, такі як морфологічні відмінності, які можуть бути важливі для деяких задач.
- Може виникнути проблема з незнайомими словами або сленгом, які не входять в словник.

```python
from tensorflow.keras.preprocessing.text import Tokenizer
num_words = 1000
# Ініціалізуємо Tokenizer
tokenizer = Tokenizer(num_words = num_words)

# Навчаємо Tokenizer на тексті
texts = ["Це перший рядок тексту.", "А це другий рядок тексту."]
tokenizer.fit_on_texts(texts)

# Перетворюємо текст на числові послідовності
sequences = tokenizer.texts_to_sequences(texts)

print("Текст:\n", texts)
print("Словник слів:", tokenizer.word_index)
print("Числові послідовності:", sequences)
```

## 2. Forward Propagation

В залежності від повнозв'язних мереж або згорткових, рекурентні будуються на векторах стану (часових векторах).

В загальному вигляді, процесс прямпого проходу виглядає наступним чином.

1. Перший етап це ініціалізація початкових ваг (векторів станів).
$$h^{(0)} = 0$$
2. Далі передача вхідних даних.
$$x^{(t)}$$
3. Обчислення активаційного значення:
$$a^{(t)} = g(W x^{(t)} + U h^{(t-1)} + b_a)$$
4. Оновлення стану.
$$h^{(t)} = a^{(t)}$$
5. Повторення кроків 2-4.
6. Виведення результату.
$$y^{(t)} = g(V h^{(t)} + b_y)$$

У цих формулах:

$x^{(t)}$ - вхідні дані на кроці часу t.

$h^{(t)}$ - стан мережі на кроці часу t.

$a^{(t)}$ - активаційне значення на кроці часу t.

$W, U, V$ - матриці ваг між вхідними даними та активаційним значенням, між попереднім станом та активаційним значенням, та між станом та вихідним значенням відповідно.

$b_a, b_y$ - зсуви.

$g$ - функція активації.

## 3. Backward Propagation (Backpropagation Through Time, BPTT)

**Backward Propagation через час (Backpropagation Through Time, BPTT)** - це алгоритм, який використовується для навчання рекурентних нейронних мереж (RNN). Він є варіантом алгоритму зворотного поширення помилки, адаптованого для рекурентних структур.

Основна ідея BPTT полягає в тому, щоб розгорнути мережу у часі на кілька кроків назад і застосувати зворотнє поширення помилки як у звичайних нейронних мережах. Це дозволяє пропускати градієнти через кожен крок часу мережі, щоб оновити ваги.

Основні кроки BPTT для навчання RNN:
1. Прямий прохід (Forward Pass)
$$a^{(t)} = g(W x^{(t)} + U h^{(t-1)} + b_a)$$
$$y^{(t)} = g(V h^{(t)} + b_y)$$
2. Обчислення функції втрат (Loss Computation).
$$L(\hat{y}, y) = \sum_{t=1}^{T} L^{(t)}(\hat{y}^{(t)}, y^{(t)})$$
3. Зворотнє поширення помилки (Backward Pass). Треба знайти похідні від $W, V, U, b_a, b_y$
- Рахуємо градієнти Лосс функції по всім вагам мережі

$$\frac{dL}{dW},\frac{dL}{dU}, \frac{dL}{dV}, \frac{dL}{db_a}, \frac{dL}{db_y} $$

$V, b_y$ можно знайти на початку
$$\frac{\partial L}{\partial V} = \sum_{t=1}^{T} \frac{\partial L^{<t>}}{\partial V}$$
$$\frac{\partial L}{\partial b_y} = \sum_{t=1}^{T} \frac{\partial L^{<t>}}{\partial b_y}$$

- Інші знаходяться як похідна віє функцій активації та активаційних значень інших станів

4. Оновлення параметрів (Parameters Update).

5. Повторення (Iteration).

## Типи архітектур рекурентних нейронних мереж

**One-to-One (Один-до-Одного):**

- В цьому випадку мережа приймає одне вхідне значення та генерує одне вихідне значення.
- Це типовий випадок для багатьох нейронних мереж, які обробляють звичайні дані, такі як класифікація зображень або передбачення тексту.

**One-to-Many (Один-до-Багатьох):**

- У цьому випадку мережа приймає одне вхідне значення і генерує послідовність вихідних значень.
- Наприклад, музичний генератор може приймати один початковий звук та генерувати музичну композицію.

**Many-to-One (Багато-до-Одного):**

- У цьому випадку мережа приймає послідовність вхідних значень і генерує одне вихідне значення.
- Наприклад, аналіз тональності тексту, де модель приймає послідовність слів і визначає загальний настрій тексту.

**Many-to-Many (Багато-до-Багатьох):**

- У цьому випадку мережа приймає послідовність вхідних значень та генерує послідовність вихідних значень.

## Задача 1. Text generation

```python
import numpy as np

from tensorflow.keras.layers import Dense, SimpleRNN, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer, text_to_word_sequence
from tensorflow.keras.utils import to_categorical

with open('/content/Передбачення_у_кав\'ярню', 'r', encoding='cp1251') as f:
    texts = f.read()
    texts = texts.replace('\ufeff', '')  # убираем первый невидимый символ
```

```python
maxWordsCount = 1000
tokenizer = Tokenizer(num_words=maxWordsCount,
                      filters='!–"—#$%&amp;()*+,-./:;<=>?@[\\]^_`{|}~\t\n\r«»',
                      lower=True,
                      split=' ',
                      char_level=False
                      )
tokenizer.fit_on_texts([texts])

dist = list(tokenizer.word_counts.items())
print(dist[:10])

data = tokenizer.texts_to_sequences([texts])
res = to_categorical(data[0], num_classes=maxWordsCount)
print(res.shape)

inp_words = 3
n = res.shape[0] - inp_words

X = np.array([res[i:i + inp_words, :] for i in range(n)])
Y = res[inp_words:]
```

https://www.tensorflow.org/api_docs/python/tf/keras/layers/SimpleRNN

```python
model = Sequential()
model.add(Input((inp_words, maxWordsCount)))
model.add(SimpleRNN(128, activation='tanh'))
model.add(Dense(maxWordsCount, activation='softmax'))
model.summary()

model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
history = model.fit(X, Y, batch_size=32, epochs=50)
```

```python
def buildPhrase(texts, str_len=10):
    res = texts
    data = tokenizer.texts_to_sequences([texts])[0]
    for i in range(str_len):
        x = to_categorical(data[i: i + inp_words], num_classes=maxWordsCount)  # преобразуем в One-Hot-encoding
        inp = x.reshape(1, inp_words, maxWordsCount)

        pred = model.predict(inp)
        indx = pred.argmax(axis=1)[0]
        data.append(indx)

        res += " " + tokenizer.index_word[indx]  # дописываем строку

    return res


res = buildPhrase("Оточіть себе позитивними")
print(res)
```

## Задача 2. Stock Price prediction

https://www.kaggle.com/code/ozkanozturk/stock-price-prediction-by-simple-rnn-and-lstm/notebook

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import os
warnings.filterwarnings("ignore")
```

### Import/initialization data

```python
# Import data
data = pd.read_csv("/content/Tesla.csv - Tesla.csv.csv")
display(data.info())
display(data.head())
```

```python
length_data = len(data)     # rows that data has
split_ratio = 0.7           # %70 train + %30 validation
length_train = round(length_data * split_ratio)
length_validation = length_data - length_train
print("Data length :", length_data)
print("Train data length :", length_train)
print("Validation data lenth :", length_validation)

# Train data
print('-'*50)
train_data = data[:length_train].iloc[:,:2]
train_data['Date'] = pd.to_datetime(train_data['Date'])  # converting to date time object
display(train_data.head())

# Val data
print('-'*50)
validation_data = data[length_train:].iloc[:,:2]
validation_data['Date'] = pd.to_datetime(validation_data['Date'])  # converting to date time object
display(validation_data.head())
```

### Preprocesing

```python
from sklearn.preprocessing import MinMaxScaler

dataset_train = train_data.Open.values
display(dataset_train.shape)
print('-'*50)
# Change 1d array to 2d array
# Changing shape from (1692,) to (1692,1)
dataset_train = np.reshape(dataset_train, (-1,1))
display(dataset_train.shape)
print('-'*50)
# scaling dataset
scaler = MinMaxScaler(feature_range = (0,1))
dataset_train_scaled = scaler.fit_transform(dataset_train)
display(dataset_train_scaled.shape)
print('-'*50)
```

```python
plt.subplots(figsize = (15,6))
plt.plot(dataset_train_scaled)
plt.xlabel("Days as 1st, 2nd, 3rd..")
plt.ylabel("Open Price")
plt.show()
```

```python
# Creating X_train and y_train from Train data
X_train = []
y_train = []

time_step = 50

for i in range(time_step, length_train):
    X_train.append(dataset_train_scaled[i-time_step:i,0])
    y_train.append(dataset_train_scaled[i,0])

# convert list to array
X_train, y_train = np.array(X_train), np.array(y_train)
print("Shape of X_train before reshape :",X_train.shape)
print("Shape of y_train before reshape :",y_train.shape)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1],1))
y_train = np.reshape(y_train, (y_train.shape[0],1))

print("Shape of X_train after reshape :",X_train.shape)
print("Shape of y_train after reshape :",y_train.shape)
```

### Modeling

```python
# importing libraries
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import SimpleRNN
from keras.layers import Dropout

# initializing the RNN
regressor = Sequential()

# adding first RNN layer and dropout regulatization
regressor.add(SimpleRNN(units = 50, activation = "tanh", return_sequences = True, input_shape = (X_train.shape[1],1)))
regressor.add(Dropout(0.2))
# adding second RNN layer and dropout regulatization
regressor.add(SimpleRNN(units = 50, activation = "tanh", return_sequences = True))
regressor.add(Dropout(0.2))
# adding third RNN layer and dropout regulatization
regressor.add(SimpleRNN(units = 50,activation = "tanh",return_sequences = True))
regressor.add(Dropout(0.2))
# adding fourth RNN layer and dropout regulatization
regressor.add(SimpleRNN(units = 50))
regressor.add(Dropout(0.2))
# adding the output layer
regressor.add(Dense(units = 1))

# compiling RNN
regressor.compile(
    optimizer = "adam",
    loss = "mean_squared_error",
    metrics = ["mae"])
regressor.summary()
```

```python
# fitting the RNN
history = regressor.fit(X_train, y_train, epochs = 50, batch_size = 32)
```

```python
history.history["loss"]
```

### Analyze of results

```python
# Plotting Loss vs Epochs
plt.figure(figsize =(10,7))
plt.plot(history.history["loss"])
plt.xlabel("Epochs")
plt.ylabel("Losses")
plt.title("Simple RNN model, Loss vs Epoch")
plt.show()
```

```python
# Plotting Accuracy vs Epochs
plt.figure(figsize =(10,5))
plt.plot(history.history["mae"])
plt.xlabel("Epochs")
plt.ylabel("Accuracies")
plt.title("Simple RNN model, Accuracy vs Epoch")
plt.show()
```

```python
y_pred = regressor.predict(X_train)  # predictions
y_pred = scaler.inverse_transform(y_pred) # scaling back from 0-1 to original
y_pred.shape
```

```python
y_train = scaler.inverse_transform(y_train) # scaling back from 0-1 to original
y_train.shape
```

```python
# visualisation
plt.figure(figsize = (30,10))
plt.plot(y_pred, color = "b", label = "y_pred" )
plt.plot(y_train, color = "g", label = "y_train")
plt.xlabel("Days")
plt.ylabel("Open price")
plt.title("Simple RNN model, Predictions with input X_train vs y_train")
plt.legend()
plt.show()
```

```python
dataset_validation = validation_data.Open.values  # getting "open" column and converting to array
dataset_validation = np.reshape(dataset_validation, (-1,1))  # converting 1D to 2D array
scaled_dataset_validation =  scaler.fit_transform(dataset_validation)  # scaling open values to between 0 and 1
print("Shape of scaled validation dataset :",scaled_dataset_validation.shape)
```

```python
# Creating X_test and y_test
X_test = []
y_test = []

for i in range(time_step, length_validation):
    X_test.append(scaled_dataset_validation[i-time_step:i,0])
    y_test.append(scaled_dataset_validation[i,0])
```

```python
# Converting to array
X_test, y_test = np.array(X_test), np.array(y_test)
print("Shape of X_test before reshape :",X_test.shape)
print("Shape of y_test before reshape :",y_test.shape)
```

```python
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))  # reshape to 3D array
y_test = np.reshape(y_test, (-1,1))  # reshape to 2D array

print("Shape of X_test after reshape :",X_test.shape)
print("Shape of y_test after reshape :",y_test.shape)
```

```python
# predictions with X_test data
y_pred_of_test = regressor.predict(X_test)
# scaling back from 0-1 to original
y_pred_of_test = scaler.inverse_transform(y_pred_of_test)
print("Shape of y_pred_of_test :",y_pred_of_test.shape)
```

```python
# visualisation
plt.figure(figsize = (30,10))
plt.plot(y_pred_of_test, label = "y_pred_of_test", c = "orange")
plt.plot(scaler.inverse_transform(y_test), label = "y_test", c = "g")
plt.xlabel("Days")
plt.ylabel("Open price")
plt.title("Simple RNN model, Prediction with input X_test vs y_test")
plt.legend()
plt.show()
```

```python
# Visualisation
plt.subplots(figsize =(30,12))
plt.plot(train_data.Date, train_data.Open, label = "train_data", color = "b")
plt.plot(validation_data.Date, validation_data.Open, label = "validation_data", color = "g")
plt.plot(train_data.Date.iloc[time_step:], y_pred, label = "y_pred", color = "r")
plt.plot(validation_data.Date.iloc[time_step:], y_pred_of_test, label = "y_pred_of_test", color = "orange")
plt.xlabel("Days")
plt.ylabel("Open price")
plt.title("Simple RNN model, Train-Validation-Prediction")
plt.legend()
plt.show()
```
